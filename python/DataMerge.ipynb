```python
#import dependencies

import pandas as pd
# settings to display all columns
pd.set_option("display.max_columns", None)
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

data_file = "Data Sets/core_dataset.csv"
data_file_two = "Data Sets/HRDataset_v9.csv"
data_file_three = "Data Sets/HRDataset_v13.csv"

employees = pd.read_csv(data_file)
employees_two = pd.read_csv(data_file_two)
employees_three = pd.read_csv(data_file_three)

employees
```

```python
employees_two
```

```python
employees_three
employees_three_new = employees_three.rename(columns={"EmpID" : "Employee Number"})
employees_three_new
```

```python
employees_df = pd.DataFrame(employees)
employees_df

employees_two_df=pd.DataFrame(employees_two)
employees_two_df

employees_three_new_df=pd.DataFrame(employees_three_new)
employees_three_new_df
```

```python
firstmerge_df =pd.merge(employees_df, employees_two_df, on="Employee Number", suffixes=('','_y'))
firstmerge_df.drop([col for col in firstmerge_df.columns if '_y'in col], axis=1, inplace=True)
firstmerge_df
```

```python
finalmerge_df =pd.merge(firstmerge_df, employees_three_new_df, on="Employee Number", suffixes=('','_y'))
finalmerge_df.drop([col for col in finalmerge_df.columns if '_y'in col], axis=1, inplace=True)
finalmerge_df.head()
```

```python
finalmerge_df.drop(["EmploymentStatus", "PayRate", "DateofHire", "DateofTermination", "TermReason", "EmploymentStatus","ManagerName","RecruitmentSource","PerformanceScore", "SpecialProjectsCount", "LastPerformanceReview_Date"], inplace=True, axis=1)
finalmerge_df
```

```python
finalmerge_df.to_csv("Data Sets/Employee Final.csv", index=False, header=True)
```
